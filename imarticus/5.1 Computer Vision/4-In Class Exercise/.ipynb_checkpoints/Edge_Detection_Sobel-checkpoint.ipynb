# EDGE DETECTION


Import libraries



```python
import cv2
import matplotlib.pyplot as plt
```

# Edge Detection using Sobel Operator

### Read the Image

```python
img = cv2.imread('Lena.tiff')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Blur the image using Gaussian Filter

```python
img = cv2.GaussianBlur(img, (3, 3), 0)
```

### Convert to Gray-Scale Image

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

### Find Image Gradient using Sobel Masks

```python
# Gradient-X
grad_x = cv2.Sobel(gray, cv2.CV_16S, 1, 0, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)

# Gradient-Y
grad_y = cv2.Sobel(gray, cv2.CV_16S, 0, 1, ksize=3, scale=1, delta=0, borderType=cv2.BORDER_DEFAULT)
    
    
abs_grad_x = cv2.convertScaleAbs(grad_x)
abs_grad_y = cv2.convertScaleAbs(grad_y)

grad = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
```

## Display Results

Let's first show the image gradients computed by the Sobel masks along the X and Y directions.

We see that these masks are able to find edges along vertical and horizontal directions.


```python
# Set a figure size of the images to be displayed
plt.figure(figsize=(15,8))

# Display the output
plt.subplot(131),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(abs_grad_x, cmap='gray'),plt.title('X_Gradient_Edgemap')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(abs_grad_y, cmap='gray'),plt.title('Y_Gradient_Edgemap')
plt.xticks([]), plt.yticks([])
```


Now, lets find the edgemap obtained by combining the gradients along both the directions.


```python

# Set a figure size of the images to be displayed
plt.figure(figsize=(15,8))

# Display the output
plt.subplot(131),plt.imshow(img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(132),plt.imshow(gray, cmap='gray'),plt.title('Gray-Scale')
plt.xticks([]), plt.yticks([])
plt.subplot(133),plt.imshow(grad, cmap='gray'),plt.title('Sobel Edge Map')
plt.xticks([]), plt.yticks([])

plt.show()
```
