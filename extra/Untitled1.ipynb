```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np 

x = np.linspace(0,5,11)
x
```

```python
y = x**3
y
```

```python
plt.plot(x,y,'v--')
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("title")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"g--")



plt.subplot(1,2,2)
plt.plot(y,x,"y--")
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,2,2])

axes.plot(x,y)

```
