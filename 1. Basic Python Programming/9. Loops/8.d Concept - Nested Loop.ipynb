# NESTED LOOPS
- Nested loops are loops that exist inside the body of another loop

```python
for i in range(1, 11):  
    for x in range(11, 21):
        for z in range(21, 31):
            for c in range(31,41):
                print(i,x,z,c)
```

```python
for x in range(1, 100): # slow loop 
    for y in range(1, 11):   # fast loop
        print('{} * {} = {}'.format(x, y, x*y))
    print("###########")
```

```python
for x in range(1, 11): # slow loop 
    for y in range(1, 11):   # fast loop
        print('{} * {} = {}'.format(x, y, x*y))
```

```python
color_list = ['black', 'white', 'gold']
phone_list = ['iphone', 'samsung', 'blackberry',]
```

```python
for i in color_list:
    for j in phone_list:
        print(i, j)
```

```python
0 black iphone
1 white samsung
2 gold iphone
```

```python
for i,(a,b) in enumerate(zip(color_list, phone_list)):
    print(i, a,b)
```

```python
for idx , (c, p) in enumerate(zip(color_list, phone_list)):
    print(idx, c, p)
```
